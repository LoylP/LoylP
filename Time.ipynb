```python
import time
import random
```

```python
# Tạo một test case ngẫu nhiên
def generate_test_case(n):
  test_case = [random.randint(1, 1000) for _ in range(n)]
  return test_case

def function(input_data):

  print(input_data)

def measure_time(n):
  test_case = generate_test_case(n)

  start_time = time.time()
  result = function(test_case)
  end_time = time.time()

  print("Test case size:", n)
  print("Execution time:", round(end_time - start_time, 10), "seconds")
```

```python
# Đo thời gian thực thi với test case có kích thước từ 1 đến 1000
for i in range(1, 50):
  measure_time(i)
```
